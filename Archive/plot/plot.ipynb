```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d1 = pd.read_csv('docu_embedding_all_feature_mse.csv')
df1 = pd.read_csv('docu_embedding_all_feature_mse 1.csv')
df1
```

```python
plt.figure(figsize = (8,5))
plt.plot(d1.year, d1.linear, color = 'y', marker = 'o', label = 'linear',alpha = 0.2)
for a, b in zip(d1.year, d1.linear):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d1.ridge, color = 'r', marker = 'o', label = 'ridge',alpha = 0.4)
for a, b in zip(d1.year,  d1.ridge):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d1.lasso, color = 'g', marker = 'o', label = 'lasso',alpha = 0.4)
for a, b in zip(d1.year,  d1.lasso):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d1.forest, color = 'b', marker = 'o', label = 'forest')
for a, b in zip(d1.year,  d1.forest):
    plt.text(a, b, b, va = 'bottom')
plt.title('Combined-feature models mse -- doc embedding')
plt.legend()
plt.xlabel('year')
plt.ylabel('MSE');
```

```python
d2 = pd.read_csv('tfidf_all_feature_mse.csv')
df2 = pd.read_csv('tfidf_all_feature_mse 1.csv')
df2
```

```python
plt.figure(figsize = (8,5))
plt.plot(d1.year, d2.linear, color = 'y', marker = 'o', label = 'linear',alpha = 0.2)
for a, b in zip(d1.year, d2.linear):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d2.ridge, color = 'r', marker = 'o', label = 'ridge',alpha = 0.4)
for a, b in zip(d1.year,  d2.ridge):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d2.lasso, color = 'g', marker = 'o', label = 'lasso',alpha = 0.4)
for a, b in zip(d1.year,  d2.lasso):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d2.forest, color = 'b', marker = 'o', label = 'forest')
for a, b in zip(d1.year,  d2.forest):
    plt.text(a, b, b, va = 'bottom')
plt.title('Combined-feature models mse -- tfidf')
plt.legend()
plt.xlabel('year')
plt.ylabel('MSE');
```

```python
same_price_mse = [11.6, 21.4, 22.2, 18.8, 11.0, 21.5, 20, 27.8, 29.1]
tfidf_mse = [35.12, 41.77, 84.86, 76.56, 50.50, 61.17, 91.48, 125.11, 166.27]
document_embedding_mse = [31.81, 48.34, 76.87, 86.43, 57.24, 61.1, 86.98, 107.59, 140.06]
price_feature = [10.75, 13.08, 12.70, 11.65, 10.44, 12.08, 14.75, 18.80, 21.10]
```

```python
plt.figure(figsize = (8,5))
plt.plot(d1.year, tfidf_mse, color = 'm', marker = 'o', label = 'tfidf_base',linestyle='--',alpha = 0.5)
for a, b in zip(d1.year, tfidf_mse):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year, document_embedding_mse, color = 'g', marker = 'o', label = 'doc_embedding_base',linestyle='--',alpha = 0.5)
for a, b in zip(d1.year, document_embedding_mse):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year, same_price_mse, color = 'y', marker = 'o', label = 'same_price_base',linestyle='--',alpha = 0.5)
for a, b in zip(d1.year, same_price_mse):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year, price_feature, color = 'b', marker = 'o', label = 'price_feature_base',linestyle='--',alpha = 0.5)
for a, b in zip(d1.year, price_feature):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year,  d1.lasso, color = 'c', marker = 'o', label = 'ridge_doc_embedding',linestyle='--',alpha = 0.5)
for a, b in zip(d1.year,  d1.lasso):
    plt.text(a, b, b, va = 'bottom')
plt.plot(d1.year, d2.forest, color = 'r', marker = 'D', label = 'forest_tfidf')
for a, b in zip(d1.year, d2.forest):
    plt.text(a, b, b, va = 'bottom')
plt.title('models mse -- baseline and best model of all feature')
plt.legend()
plt.xlabel('year')
plt.ylabel('MSE');
```

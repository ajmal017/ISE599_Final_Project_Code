```python
Symbol_Name_NASDAQ = pd.DataFrame(columns = ['Symbol','Company_name'])
File_not_separate_NASDAQ = pd.DataFrame(columns = ['Row_#','String'])
```

```python
index_c = 0
index_n = 0
for row in range(0,len(file_2)):
    if row != 0 and row != len(file_2)-1:

        if ' -' in file_2[row]:
            sen = re.match("(.*?) -",file_2[row]).group()
            sen_2 = sen[:sen.index(" -")]
            sen_3 = sen_2.split('|')
            Symbol_Name_NASDAQ.loc[index_c,'Symbol'] = sen_3[0]
            Symbol_Name_NASDAQ.loc[index_c,'Company_name'] = sen_3[1]
            index_c +=1
        else:
            print ('This stock did not match:\n',row,' ',file_2[row])
            File_not_separate_NASDAQ.loc[index_n,'Row_#'] = row
            File_not_separate_NASDAQ.loc[index_n,'Row_#'] = file_2[row]
            index_n +=1
```

```python
import re, requests

def getCIKs(TICKERS):
    URL = 'http://www.sec.gov/cgi-bin/browse-edgar?CIK={}&Find=Search&owner=exclude&action=getcompany'
    CIK_RE = re.compile(r'.*CIK=(\d{10}).*')    
    cik_dict = {}
    for ticker in TICKERS:
        f = requests.get(URL.format(ticker), stream = True)
        results = CIK_RE.findall(f.text)
        if len(results):
            results[0] = int(re.sub('\.[0]*', '.', results[0]))
            cik_dict[str(ticker).upper()] = str(results[0])
    f = open('cik_dict', 'w')   
    return cik_dict
    f.close()
```

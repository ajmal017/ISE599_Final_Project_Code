Since each 10k file only had unique CIK number, and the avg stock price features is identified by symbol(Ticker), we will need to first convert the stock prices symbol to CIK and then map them to the CIK contents.

1) Select the 10k file that its CIK can be found in CIK_Symbol files.
2) Find the intersection between 10k files and stock prices features.
3) Create dict for the each CIK and its corresponded symbol(Ticker)
4) Map symbol to CIK
5) Use CIK as identification to merge 10 contents and stock price features.

```python
def Dataframe_merge(CIK_Symbol, CIK_10K, Symbol_price):
    
    CIK_in_map = CIK_Symbol[CIK_Symbol['CIK'].isin(CIK_10K['CENTRAL INDEX KEY:'])]
    
    symbol_map = {}
    
    for row in range(0,len(CIK_in_map)):
        symbol_map[CIK_in_map.iloc[row,1]] = CIK_in_map.iloc[row,0]
        
    CIK_symbol_in_map = Symbol_price[Symbol_price['symbol'].isin(CIK_in_map['Ticker'])]
    CIK_symbol_in_map['symbol'] = CIK_symbol_in_map['symbol'].map(symbol_map)
    
    Merge_dataframe = CIK_10K.merge(CIK_symbol_in_map, left_on ='CENTRAL INDEX KEY:', right_on = 'symbol' )
    
    return Merge_dataframe
```
